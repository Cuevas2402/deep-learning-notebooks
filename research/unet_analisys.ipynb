```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rdf1 = pd.read_csv('./db/r1.csv')
rdf2 = pd.read_csv('./db/r2.csv')
rdf3 = pd.read_csv('./db/r3.csv')

rdf = pd.concat([rdf1, rdf2, rdf3], ignore_index=True)
```

```python
rdf
```

```python
udf = pd.read_csv('./db/u1.csv')
```

```python
udf
```

```python
udf.columns
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))


ax1.plot(udf['Train Loss'], label='U-Net', color='b', marker='o')
ax1.plot(rdf['Train Loss'], label='ResU-Net', color='r', marker='o')
ax1.set_title('Pérdida en Entrenamiento')
ax1.set_xlabel('Pérdida')
ax1.set_ylabel('Época')
ax1.legend()


ax2.plot(udf['Val Loss'], label='U-Net', color='b', marker='o')
ax2.plot(rdf['Val Loss'], label='ResU-Net', color='r', marker='o')
ax2.set_title('Pérdida en Validación')
ax2.set_xlabel('Pérdida')
ax2.set_ylabel('Época')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(2, 3, figsize=(12, 8))


axes[0, 0].plot(udf['Accuracy'], label='U-Net', color='b', marker='o')
axes[0, 0].plot(rdf['Accuracy'], label='ResU-Net', color='r', marker='o')
axes[0, 0].set_title('Accuracy')
axes[0, 0].legend()

axes[0, 1].plot(udf['Precision'], label='U-Net', color='b', marker='o')
axes[0, 1].plot(rdf['Precision'], label='ResU-Net', color='r', marker='o')
axes[0, 1].set_title('Precision')
axes[0, 1].legend()

axes[0, 2].plot(udf['Recall'], label='U-Net', color='b', marker='o')
axes[0, 2].plot(rdf['Recall'], label='ResU-Net', color='r', marker='o')
axes[0, 2].set_title('Recall')
axes[0, 2].legend()

axes[1, 0].plot(udf['F1 Score'], label='U-Net', color='b', marker='o')
axes[1, 0].plot(rdf['F1 Score'], label='ResU-Net', color='r', marker='o')
axes[1, 0].set_title('F1 Score')
axes[1, 0].legend()

axes[1, 1].plot(udf['Dice Score'], label='U-Net', color='b', marker='o')
axes[1, 1].plot(rdf['Dice Score'], label='ResU-Net', color='r', marker='o')
axes[1, 1].set_title('Dice Score')
axes[1, 1].legend()

fig.delaxes(axes[1, 2])

plt.tight_layout()
plt.show()
```
